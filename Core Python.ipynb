```python
#7)Private varieble - Private varieble can't access outside of class,Only accessseble inside of class with help of (class.private_varieble_name)
class A:
  __a = 10 #Private varieble
  def __init__(self):
    print(self.__a) #accesseble inside of constractor
  def test(self):
    print(self.__a) #accesseble inside of class

a = A()
a.test()
print(a.__a)#Private varieble can't access outside of class
```

#2)Encapsulation

###1)Why we use encapsulation.
###2)Private aatributes.
###3)Getter Seeter.

```python
#1)Why we use Encapsualtion -
""" 1)We when we create varieble and methods inside a class we don't want to show the user or outside--->show we made variebles as private so can't access outside of class but will be accessible inside a class.
    2)But other programmer want to use this varieble so give permission to use with certain conditions on variebles--->putting ceratian condition for varieble so programmer can't do measure changes we create function are called getter and setter. """

class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute
        self.__age = age    # Private attribute

    # Public method to get the name
    def get_name(self):
        return self.__name

    # Public method to set the name
    def set_name(self, name):
      if name== str:
        self.__name = name
      else:
        print("Invalid name")

    # Public method to get the age
    def get_age(self):
        return self.__age

    # Public method to set the age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age")
# Creating an instance of the Person class
person = Person("Alice", 30)

# Accessing the private attributes via public methods
print(person.get_name())  # Output: Alice
print(person.get_age())   # Output: 30

# Modifying the private attributes via public methods
person.set_name("Bob")
person.set_age(25)

print(person.get_name())  # Output: Bob
print(person.get_age())   # Output: 25
```

###2)inheritance

```python
#)Aggregation-
"""when we bring data from 1st class instance varieble to 2nd class instance varieble is called aggregation."""
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print("Engine started with horsepower:", self.horsepower)

class Car:
    def __init__(self, brand, engine2):
        self.brand = brand
        self.engine3 = engine2  # now engine3 which is refernce is holding the value of reference varieble engine1

    def drive(self):
        print(f"{self.brand} car is driving.")
        self.engine3.start()

# Creating an Engine object
engine1 = Engine(horsepower=150)

# Creating a Car object and passing the Engine object to it
car = Car(brand="Toyota", engine2=engine1)

# Using the Car object
car.drive()
```
